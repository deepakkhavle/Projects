```python
# To get information of bill

def bill_information():
    clear() 
    bill_no = input('Enter Bill Number :') # Enter bill no as per requirement
                                                              
                                           # To get records from backend database
    sql = F"""select b.id, b.name, b.phone, b.bill_date, t.item_id,t.qty,i.item_name, 
           i.price from bills b,transactions t,items i 
           where b.id = t.bill_id AND t.item_id= i.id AND 
           b.id ={bill_no};"""

    cursor.execute(sql) 

    records = cursor.fetchall() # To get all record from desired query

    n = cursor.rowcount # counting the generated records

    clear()

    print("Bill No :",bill_no) 
    print('-'*100)

    if n<=0: # If entered value is less than or equal to zero then print bill no not exist
        print(f'Bill number {bill_no} does not exists')

    else:                                                               #shows records as per entered bill no
        print(f'Customer Name : {records[0][1]}  Phone No :{records[0][2]}') 

        print(f'Bill Date : {records[0][3]}'.format(records[0][3]))

        print('-'*100)

        print('{:10s} {:30s} {:20s} {:10s} {:30s}'.format('ID','Item Name','Qty', 
             'Price','Amount'))
        print('-'*100)

        for row in records:
            print('{:<10d} {:30s} {:<20d} {:.2f}    {:>.2f}'.format(row[4],row[6],row[5],row[7],row[5]*row[7]))

        print('-'*100)

    print('\nPress any key to continue....')
    wait = input()
```

```python
# To get amount collected between two date

def amount_collected():
    clear()
    
    start_date = input('Enter start Date (YYYY-MM-DD) :') # Enter  start date as per owner's requirement
    end_date   = input('Enter End   Date (YYYY-MM-DD) :') # Enter  end date as per owner's requirement
    clear()

    print(f'Amount collected between: {start_date} and {end_date}')
    print('-'*100)
                                                                             # To get data from backend
    sql = f"""select round(sum(t.qty*i.price),2) from bills b,transactions t,items i \           
           where b.bill_date between "{start_date}" AND "{end_date}" AND b.id = t.bill_id AND \
           t.item_id = i.id;"""
    cursor.execute(sql)
    result = cursor.fetchone()[0]

    print(result)

    print('\nPress any key to continue.....')
    wait= input()
```
